```python
# ============================================
# Check extracted folder
# ============================================

import os

os.listdir("northstar_dataset/northstar_dataset")
```

```python
# ============================================
# Load Datasets
# ============================================

import pandas as pd

base_path = "northstar_dataset/northstar_dataset"

customers = pd.read_csv(f"{base_path}/customers.csv")
orders = pd.read_csv(f"{base_path}/orders.csv")
deliveries = pd.read_csv(f"{base_path}/deliveries.csv")
drivers = pd.read_csv(f"{base_path}/drivers.csv")
vehicles = pd.read_csv(f"{base_path}/vehicles.csv")
hubs = pd.read_csv(f"{base_path}/hubs.csv")
incidents = pd.read_csv(f"{base_path}/incidents.csv")
complaints = pd.read_csv(f"{base_path}/complaints.csv")
app_events = pd.read_csv(f"{base_path}/app_events.csv")

print("All datasets loaded successfully!")
```

```python
# ============================================
# STORE DATASETS IN DICTIONARY
# ============================================

datasets = {
    "customers": customers,
    "orders": orders,
    "deliveries": deliveries,
    "drivers": drivers,
    "vehicles": vehicles,
    "hubs": hubs,
    "incidents": incidents,
    "complaints": complaints,
    "app_events": app_events
}
```

```python
# ============================================
# DUPLICATE RECORD ANALYSIS
# ============================================

print("DUPLICATE RECORD ANALYSIS\n")

duplicate_summary = []

for name, df in datasets.items():

    duplicate_count = df.duplicated().sum()

    duplicate_summary.append({
        "Dataset": name,
        "Duplicate Records": duplicate_count
    })

    print("\n============================")
    print(name.upper())
    print("============================")
    print(f"Duplicate Records: {duplicate_count}")
```

```python
# ============================================
# DUPLICATE SUMMARY TABLE
# ============================================

duplicate_df = pd.DataFrame(duplicate_summary)

duplicate_df
```

```python
# ============================================
# VISUALISATION
# ============================================

plt.figure(figsize=(10,6))

sns.barplot(
    data=duplicate_df,
    x="Dataset",
    y="Duplicate Records"
)

plt.title("Duplicate Records Across NorthStar Datasets")
plt.xlabel("Dataset")
plt.ylabel("Number of Duplicate Records")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

### Interpretation

The duplicate record analysis across all NorthStar datasets shows that no duplicate rows were detected in any table. This indicates that each entity (customers, orders, deliveries, drivers, vehicles, hubs, incidents, complaints, and app_events) maintains a unique record structure.

This is a positive outcome for data integrity, as it ensures that there is no redundant data that could distort aggregation results, SQL joins, or statistical analysis in later chapters. It also confirms that primary key constraints (e.g., customer_id, order_id, delivery_id) are properly maintained at the dataset level.
